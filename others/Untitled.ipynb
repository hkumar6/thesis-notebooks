## Bifurcation functions on the consensus manifold


### Saddle-node

$S_{sn}(x,u) = ue^{(x+1)} - 1 $

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#plt.figure(figsize=(16,9))
x = np.arange(-1,1,0.001)
#plt.subplot(1,3,1)
u = np.exp(-1) - 0.05
plt.plot(x, u*np.exp(x+1)-1, label=r'$u < u^*$')

#plt.subplot(1,3,2)
u = np.exp(-1)
plt.plot(x, u*np.exp(x+1)-1, label=r'$u = u^*$')

#plt.subplot(1,3,3)
u = np.exp(-1) + 0.05
plt.plot(x, u*np.exp(x+1)-1, label=r'$u > u^*$')
plt.plot(x,x,color="black", label=r'$x = y$')

plt.legend()
plt.savefig("saddle_node_fn.pdf")
```

### Transcritical

```python
x = np.arange(-0.5,0.5,0.001)
s_tr = lambda x: np.square(x) + u*x

u = 0.5
plt.plot(x, s_tr(x), label=r'$u < u^*$')
u = 1
plt.plot(x, s_tr(x), label=r'$u = u^*$')
u = 1.5
plt.plot(x, s_tr(x), label=r'$u > u^*$')

plt.plot(x,x,color="black", label=r'$x = y$')

plt.legend()
plt.savefig("transcritical_fn.pdf")
```

### Pitchfork

```python
x = np.arange(-2,2,0.001)
s_tr = lambda x: u*np.tanh(x)

u = 0.5
plt.plot(x, s_tr(x), label=r'$u < u^*$')
u = 1
plt.plot(x, s_tr(x), label=r'$u = u^*$')
u = 1.5
plt.plot(x, s_tr(x), label=r'$u > u^*$')

plt.plot(x,x,color="black", label=r'$x = y$')

plt.legend()
plt.savefig("pitchfork_fn.pdf")
```
